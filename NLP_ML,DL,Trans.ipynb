```python
import pandas as pd
import numpy as np

# Simple real-world style dataset
data = {
    "text": [
        "The app keeps crashing",
        "Customer support was very helpful",
        "Worst experience ever",
        "Loved the new update",
        "Not satisfied with the service",
        "Excellent product quality",
        "Terrible response from support",
        "Very happy with the purchase"
    ],
    "label": [0, 1, 0, 1, 0, 1, 0, 1]  # 0 = Negative, 1 = Positive
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["label"], test_size=0.25, random_state=42
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

ml_model = LogisticRegression()
ml_model.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

ml_pred = ml_model.predict(X_test_vec)

print("ML Accuracy:", accuracy_score(y_test, ml_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, ml_pred))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(df["text"])

sequences = tokenizer.texts_to_sequences(df["text"])
X_dl = pad_sequences(sequences, maxlen=10)
y_dl = df["label"].values
```

```python
X_train_dl, X_test_dl, y_train_dl, y_test_dl = train_test_split(
    X_dl, y_dl, test_size=0.25, random_state=42
)
```

```python
dl_model = tf.keras.Sequential([
    tf.keras.layers.Embedding(5000, 64, input_length=10),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

dl_model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

dl_model.summary()
```

```python
dl_model.fit(
    X_train_dl, y_train_dl,
    epochs=10,
    batch_size=2,
    verbose=1
)
```

```python
loss, acc = dl_model.evaluate(X_test_dl, y_test_dl)
print("DL Accuracy:", acc)
```

```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
texts = df["text"].tolist()

results = classifier(texts)

for t, r in zip(texts, results):
    print(f"{t} --> {r}")
```
